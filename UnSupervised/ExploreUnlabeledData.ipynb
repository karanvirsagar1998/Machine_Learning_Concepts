```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('Mall_Customers.csv')

# Quick look at the data
data.head()
```

```python
data.describe()
```

```python
data.info()
```

Preprocess the Data

```python
data.isnull().sum()
```

```python
data.rename(columns={"Annual Income (k$)" :"Annual Income", "Spending Score (1-100)":"Spending Score"}, inplace=True)
```

 Visualize Basic Distributions

```python
# Visualize distributions using histograms and density plots
# Age
plt.figure()
plt.hist(data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

plt.figure()
plt.hist(data['Annual Income'], bins=20)
plt.xlabel('Income')
plt.ylabel('Distribution')
plt.show()

plt.figure()
plt.hist(data['Spending Score'], bins=20)
plt.xlabel('Spending')
plt.ylabel('Distribution')
plt.show()

# Optional: Density plot for Age sns.kdeplot(data['Age'], fill=True) # Added KDE plot example plt.title('Age Distribution (KDE)') # Added (KDE) for clarity plt.xlabel('Age') plt.ylabel('Density') plt.show() 
```

# Explore relationships between features with scatter plots

```python
plt.figure()
plt.scatter(data['Annual Income'], data['Spending Score'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()

sns.pairplot(data[['Annual Income', 'Spending Score']])
plt.show()
```
